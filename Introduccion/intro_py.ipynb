## Histograma
Por columnas de varios datos en cantidades bien distribuidas permite entender los datos a partir de gr√°ficas de una variable (frecuencias). El histograma agrupa valores en "bins" (intervalos) y muestra la altura proporcional a la cantidad de observaciones en cada bin. ### Distribuciones Bimodales
