## Propuesto

```python
ax = df.plot.bar(
    x='dataset',
    y=['algoritmo1','algoritmo2'],
    title='Evaluacion de algoritmo1 vs algoritmo2',
    figsize=(12,5),
    xlabel='Dataset',
    ylabel='Rendimiento',
)
ax.grid(
    visible=True,
    alpha=0.3,
)
ax.legend(
    labels=['Algoritmo 1','Algoritmo 2'],
    loc='upper left',
)
```

## Solucion

```python
ax = df.plot(
    kind='bar',
    x='dataset',
    y=['algoritmo1', 'algoritmo2'],
    figsize=(12, 5)
)

ax.grid(visible=True)
ax.set_xlabel("Dataset")
ax.set_ylabel("Rendimiento")
ax.set_title("Evaluacin de Algoritmo1 vs Algoritmo2")
ax.legend(labels=['Algoritmo 1', 'Algoritmo 2'], loc='upper left', bbox_to_anchor=(0, 1))
ax.set_axisbelow(True) # Ensure grid lines are behind bars
```

```python
import seaborn as sns
```

```python
X = ['Group A','Group B','Group C','Group D']
Ygirls = [10,20,20,40]
Zboys = [20,30,25,30]
  
X_axis = np.arange(len(X))
plt.bar(X_axis - 0.2, Ygirls, 0.4, label = 'Girls')
plt.bar(X_axis + 0.2, Zboys, 0.4, label = 'Boys')
plt.xticks(X_axis, X)
plt.xlabel("Groups")
plt.ylabel("Number of Students")
plt.legend()
plt.show()
```
