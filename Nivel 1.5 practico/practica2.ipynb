##  Grafico de Barras Aplicadas

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('../data/datasets_top_4.csv', index_col=0)
df
```

## Propuesto

```python
fig, ax = plt.subplots(1,2)
# separar en 2 partes para mostrar la info en 2 graficas
df1 = df.iloc[:14:] # primeras 14 filas
df2 = df.iloc[14::] # filas restantes
# grafico 1
df1.plot.barh(
    stacked=True,
    ax=ax[0],
    width=0.4,
)
# grafico 2
df2.plot(
    kind='barh',
    stacked=True,
    ax=ax[1],
    width=0.4,
)
# arreglos
fig.suptitle(
    'Como se distribuyen las categorias\ndentro de los datasets',
    fontweight='bold'
)
ax[0].set_xlabel('Porcentaje')
ax[1].set_xlabel('Porcentaje')
ax[0].set_ylabel('Datasets')
ax[1].set_ylabel('')
ax[1].yaxis.tick_right()

ax[0].legend(
    labels=['Categoria 1','Categoria 2','Categoria 3','Otros'],
    loc='upper center',
    bbox_to_anchor=(1.1, -0.12),
    ncol=4,
)
ax[1].legend_.remove()
```

## Solucion

```python
df1 = df.iloc[:14,:]
df2 = df.iloc[14:,:]
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2)

fig.set_size_inches(7, 4)

df1.plot.barh(
    stacked=True,
    ax=ax1,
    width=0.4,
    color={
        'cat1': '#000086',
        'cat2': '#5396f4',
        'cat3': '#a1dbe8',
        'otros': '#DADADA'
    }
)

df2.plot.barh(
    stacked=True,
    ax=ax2,
    width=0.4,
    color={
        'cat1': '#000086',
        'cat2': '#5396f4',
        'cat3': '#a1dbe8',
        'otros': '#DADADA'
    }
)

ax2.set_ylabel('')
ax1.set_ylabel('Dataset')

ax2.set_xlabel('Porcentaje')
ax1.set_xlabel('Porcentaje')

ax1.legend().remove()

ax2.legend(
    ['Categoria 1', 'Categoria 2', 'Categoria 3', 'Otros'],
    ncol=4,
    loc="upper left",
    bbox_to_anchor=(-1.3, -0.15)
)

ax2.yaxis.tick_right()

ax1.text(
    75, -5,
    "Algunos datasets han sido ignorados del analisis",
    fontsize=6,
    bbox={
        "facecolor": "orange",
        "alpha": 0.5
    }
)

fig.suptitle('Como se distribuyen las categorias\ndentro de los datasets?')
```
